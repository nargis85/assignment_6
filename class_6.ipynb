# encoding

```python
#label encoding
from sklearn.preprocessing import LabelEncoder #importing labelencoder for making string to number
le= LabelEncoder()
df.head()
```

```python
df.sex=le.fit_transform(df[['sex']])
df.head()
```

```python
df.sex=le.fit_transform(df[['smoker']])
df.head()
```

```python
#doing this by loop
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    else:
        df[column]=le.fit_transform(df[column])
```

```python
df.head() #with loop we made all string encode
```

```python
from pandas.core.dtypes.common import is_numeric_dtype  #thre others way to make data encoding
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    df3[col]=le.fit_transform(df3[col])    
```

# one-hot encoding

```python
df4.head()
```

```python
dummies = pd.get_dummies(df4['sex']) #one hot encoder do column for each value
dummies
```

```python
merged = pd.concat([df4,dummies],axis='columns') #merged dummies with the table
merged
```

```python
df5=merged.drop(['sex'],axis='columns') # deleted sex column as its not need cause it is divided by one-hot encoding
df5
```

```python
dummies=pd.get_dummies(df5['smoker'])
dummies
```

```python
merged=pd.concat([df5,dummies],axis='columns') #merged dummies with the table
merged
```

```python
df6=merged.drop(['smoker'],axis='columns') # deleted sex column as its not need cause it is divided by one-hot encoding
df6
```

```python
dummies=pd.get_dummies(df6['region'])
dummies
```

```python
merged=pd.concat([df6,dummies],axis='columns')
merged
```

```python
df7=merged.drop(['region'],axis='columns')
df7
```

```python
y=df7['charges'] #charges in dependent variable so kept that in y axis
y #label column
```

```python
x=df7.drop(['charges'],axis=1) #except charges other is dependent variable so we dropped charges and other value come in x axis
x
```

```python
from sklearn.linear_model import LinearRegression #
model=LinearRegression() #object create
```

```python
model.fit(x,y) #training by dataset
```

```python
model.score(x,y)
```

```python
model.predict([[20,29,1,1,1,1,1,0,1,0,1]]) # predict value
```
